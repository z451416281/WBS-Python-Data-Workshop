### Using masks to filter elements

Create two masks on our main dataframe `df`, one for all the frequencies and one for the sound pressure values. Using these masks get the dataframes with:

* Elements with frequencies higher than 1500 Hz
* Elements with sound pressure values lower than 126.0
* Elements where both of these conditions are true
* Elements where either of these conditions are true

````python
# add your code here
freq_mask = df['Frequency [Hz]'] > 1500
sound_mask = df['Sound pressure [dB]'] < 126.0

print(df.loc[freq_mask].head())
print(df.loc[sound_mask].head())
print(df.loc[freq_mask & sound_mask].head())
print(df.loc[freq_mask | sound_mask].head())
````

### Working with a pd.Series

Retrieve the series corresponding to the sound pressure from the dataframe, display

* show the name of the series (it should be `Sound pressure [dB]`)
* show the shape of the series (it should be `(1503,)`) 
* the mean and the median (resp. `124.84` and `125.72`)
* the mean of the squared values (it should be `15631.57`)
* Check [the documentation](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.html) and try some other methods and attributes

````python
# add your code here
sp = df['Sound pressure [dB]']
print("Name: {}".format(sp.name))
print("Shape: {}".format(sp.shape))
print("Mean: {0:.2f}, Median: {1:.2f}".format(sp.mean(), sp.median()))
print("Mean of sq. vals: {0:.2f}".format((sp**2).mean()))
````

## More useful tricks with Pandas

Join the column `Frequency [Hz]` and `df2` using `concat`, call the result `df3` and display the summary statistics.

````python
# add your code here to create df3
df3 = pd.concat([df['Frequency [Hz]'], df2], axis=1)
df3.describe([0.1, 0.9])
````

### Groupby and apply

Group the rows of `df3` by Frequency and call the result `freq_df3`, show the length of this object, does it correspond to what you expected?

Then:
* Using directly `.mean()` on your groupby object, compute the mean of each features per group of frequencies.
* Using `apply`, compute the max of each features per group of frequencies (passing the `max` function to `apply`)
* Using `agg`, compute the mean and the max of the Chord length and the FS velocity

Tip: `apply` and `agg` expect you to pass the functions you want to apply on the groupby object. Instead of implementing those functions (mean and max) yourself, you can import the library `numpy` that has them already available.

````python
# add your code here to compute freq_df3
freq_df3 = df3.groupby(by='Frequency [Hz]')
len(freq_df3)
````

````python
# add your code here to compute the means
import numpy as np

freq_df3.mean()
freq_df3.apply(np.max)
freq_df3.agg([np.max, np.min, np.sum, np.mean])
````

### (Bonus) Joining dataframes
